```python
def filling_median_numberical_columns(data):
    for col in data.columns:
        if data[col].dtype != 'O':
            data[col]=data[col].fillna(data[col].median())
```

```python
def combine_splitted_columns(data):
    combine_list=['Product_Info','Employment_Info','InsuredInfo','Insurance_History','Family_Hist','Medical_History','Medical_Keyword']
    for c in combine_list:
        col_list=[]
        for col in data.columns:
            if c in col and data[col].dtype != 'O':
                col_list.append(col)
    
        if len(col_list)!=0:
            data.loc[:,(c)]=0
            for col in col_list:
                data[c]=data[col]+data[c]
                data.drop(col,axis=1,inplace=True)
```

```python
def detect_outliers(data):
    for col in data.columns:
        if data[col].dtype != 'O':
            percentile_25=np.percentile(data[col],25)
            percentile_75=np.percentile(data[col],75)
            IQR=percentile_75-percentile_25
            upper_bound=percentile_75 + IQR * 1.25
            lower_bound=percentile_25 - IQR * 1.25
            print(col,":",data[data[col] > upper_bound].shape[0] + data[data[col] < lower_bound].shape[0])
```

```python
def check_missing_percentage(data,threshold=0.1):
    #Columns with missing value greater than 10 %
    columns_names=[]
    total_rows=data.shape[0]
    for col in data.columns:
        missing_values=data[col].isna().sum()
        percentage_missing=missing_values/total_rows
        if percentage_missing > threshold:
            columns_names.append(col)
            print("Missing Value Count for ",col,"=",missing_values)
            print("% Missing Value for ",col,"=",percentage_missing)
    print("Columns with missing values =",columns_names)
```

<h3> Visualization to eleminate columns</h3>

```python
def plot_heatmap(data):
    figure_name = plt.figure(figsize=(12, 8))
    sns.heatmap(data.corr(),cmap='Blues',annot=True)
```

```python
def distribution_curves(data):
    numerical_col=[col for col in data.columns if data[col].dtype != 'O']
    n_rows=int(len(numerical_col)/2)
    n_cols=2
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols,figsize=(12, 25))
    for i, column in enumerate(numerical_col):
        sns.histplot(data[column],ax=axes[i//n_cols,i%n_cols],kde=True)
```

Encoding Product_Info_2 to numerical values based on the first alphabet. Ax-Az = 1, Bx-Bz=2 .... 
<br>Reducing the variablity in Medical_Keyword

```python
def transform_Product_Info_2(data):
    data.loc[data.Product_Info_2.isin(['A1','A2','A3','A4','A5','A6','A7','A8']),'Product_Info_2']=0
    data.loc[data.Product_Info_2.isin(['B1','B2']),'Product_Info_2']=1
    data.loc[data.Product_Info_2.isin(['C1','C2','C3','C4']),'Product_Info_2']=2
    data.loc[data.Product_Info_2.isin(['D1','D2','D3','D4']),'Product_Info_2']=3
    data.loc[data.Product_Info_2.isin(['E1']),'Product_Info_2']=4
```

```python
def transform_Medical_Keyword(data):
    data.loc[data.Medical_Keyword.isin([i for i in range(1,6)]),'Medical_Keyword']=1
    data.loc[data.Medical_Keyword.isin([i for i in range(6,10)]),'Medical_Keyword']=2
    data.loc[data.Medical_Keyword.isin([i for i in range(10,17)]),'Medical_Keyword']=3
```

```python
# Create StratifiedKFold object.
def Stratifed_KFold_Train(data_X,data_y,model):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
    lst_accu_stratified = []
  
    for train_index, test_index in skf.split(data_X, data_y):
        x_train_fold, x_test_fold = data_X.iloc[train_index], data_X.iloc[test_index]
        y_train_fold, y_test_fold = data_y.iloc[train_index], data_y.iloc[test_index]
        model.fit(x_train_fold, y_train_fold.values.ravel())
        lst_accu_stratified.append(balanced_accuracy_score(model.predict(x_test_fold), y_test_fold))
    return lst_accu_stratified
```
